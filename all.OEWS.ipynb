```python

import pandas as pd
```

## Step 1:Extract the all_data_M_2023.xlsx

```python
# Load the data into a pandas Dataframe
all_oews_df= pd.read_excel("Data/all_data_M_2023.xlsx")
all_oews_df.head()
```

## Step 2:Transform the data

```python
# Area type:1=U.S:2=State 3=U.S. Territory:4 Metropolitant Statistical Area (MSA) or New England
# Narrow down the data to State level
state_oews_df = all_oews_df.loc[all_oews_df['AREA_TYPE']==2,:]
state_oews_df
```

```python
#get rid of unneccery column

clean_oews_df = state_oews_df[['AREA', 'AREA_TITLE', 'PRIM_STATE','OCC_CODE', 'OCC_TITLE', 'O_GROUP',\
     'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT','H_MEAN','A_MEAN', 'MEAN_PRSE', 'H_PCT25', 'H_MEDIAN', 'H_PCT75','A_PCT25', 'A_MEDIAN', 'A_PCT75']]
clean_oews_df
```

```python
#rename columns for better understanding
#df.rename(columns=("old name":"new name","old name":"new name"))
renamed_oews_df = clean_oews_df.rename(columns={
    'AREA': 'area',
    'AREA_TITLE': 'area_title',
    'PRIM_STATE': 'prim_state',
    'OCC_CODE': 'occ_code',
    'OCC_TITLE': 'occ_title',
    'O_GROUP': 'o_group',
    'TOT_EMP': 'tot_emp',
    'EMP_PRSE': 'emp_prse',
    'JOBS_1000': 'jobs_1000',
    'LOC_QUOTIENT': 'loc_quotient',
    'H_MEAN': 'h_mean',  
    'A_MEAN': 'a_mean',  
    'MEAN_PRSE': 'mean_prse',
    'H_PCT25': 'h_pct25',
    'H_MEDIAN': 'h_median',
    'H_PCT75': 'h_pct75',
    'A_PCT25': 'a_pct25',  
    'A_MEDIAN': 'a_median',
    'A_PCT75': 'a_pct75'
})

renamed_oews_df.head()
```

```python
# Check the data data type and non-null count (preparation for transformation)
renamed_oews_df.info()
```

```python
# Convert object columns to numeric
columns_to_convert = [
    'tot_emp', 
    'emp_prse', 
    'jobs_1000', 
    'loc_quotient', 
    'h_mean', 
    'a_mean', 
    'mean_prse', 
    'h_pct25', 
    'h_median', 
    'h_pct75', 
    'a_pct25', 
    'a_median', 
    'a_pct75'
]
```

```python
# Convert the specified columns to numeric, coercing errors
renamed_oews_df[columns_to_convert] = renamed_oews_df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

renamed_oews_df.info()
```

```python
group_total = renamed_oews_df.loc[renamed_oews_df['o_group'] == 'total', :]
group_total_count = len(group_total['occ_title'].unique())

group_major = renamed_oews_df.loc[renamed_oews_df['o_group'] == 'major', :]
group_major_count = len(group_major['occ_title'].unique())

group_detailed = renamed_oews_df.loc[renamed_oews_df['o_group'] == 'detailed', :]
group_detailed_count = len(group_detailed['occ_title'].unique())

# Corrected print statements
print(f'There are {group_total_count} unique values under group total.')
print(f'There are {group_major_count} unique values under group major.')
print(f'There are {group_detailed_count} unique values under group detailed.')
```

```python
occ_code_list = ['00-0000', '11-0000', '13-0000', '15-0000', '17-0000', '19-0000', '21-0000', '23-0000', '25-0000', '27-0000', '29-0000', '31-0000', '33-0000',\
    '35-0000', '37-0000', '39-0000', '41-0000', '43-0000', '45-0000', '47-0000', '49-0000', '51-0000', '53-0000', '11-3021', '13-1161', '13-2031', '13-2041',\
        '13-2054', '15-1221', '15-1251', '15-1252', '15-1253', '15-1254', '15-2011', '15-2031', '15-2041', '19-1022', '19-1029', '19-2041', '29-9091', '41-3021',\
            '43-9021', '13-2051','15-1211', '15-1242', '15-1243', '15-2051']

df_filtered = renamed_oews_df[renamed_oews_df['occ_code'].isin(occ_code_list)]
df_filtered
```

## Step 3: Load the Data


```python
# Export the data to CSV
df_filtered.to_csv('data/EmploymentWageData.csv', index=False)
```

```python
industry_df = df_filtered[['area', 'area_title', 'prim_state']].drop_duplicates()
```

```python
occupation_df = df_filtered[['occ_code', 'occ_title', 'o_group']].drop_duplicates()
```

```python
employment_df = df_filtered[['occ_code', 'tot_emp', 'emp_prse', 'jobs_1000', 'loc_quotient', 'h_mean', 'a_mean', 'mean_prse']].drop_duplicates()
```

```python
# Save Industry Data to CSV
industry_df.to_csv('industry_data.csv', index=False)

# Save Occupation Data to CSV
occupation_df.to_csv('occupation_data.csv', index=False)

# Save Employment Data to CSV
employment_df.to_csv('employment_data.csv', index=False)
```
